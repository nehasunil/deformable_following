```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def imshow(img):
    plt.imshow(img[:,:,::-1])
    plt.show()
    
# img = cv2.imread("fabric0.jpg")
img = cv2.imread("data/imgs/0503/color_20.jpg")
imshow(img)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_green = np.array([40,20,20])
upper_green = np.array([80,255,255])
lower_red_1 = np.array([0,20,20])
upper_red_1 = np.array([10,255,255])
lower_red_2 = np.array([170,20,20])
upper_red_2 = np.array([180,255,255])
lower_yellow = np.array([20,100,100])
upper_yellow = np.array([40,255,255])

mask_green = cv2.inRange(img_hsv, lower_green, upper_green)
mask_red = cv2.inRange(img_hsv, lower_red_1, upper_red_1) | cv2.inRange(img_hsv, lower_red_2, upper_red_2)
mask_yellow = cv2.inRange(img_hsv, lower_yellow, upper_yellow)

# plt.imshow(mask_yellow)
# plt.show()

# plt.imshow(mask_red)
# plt.show()

# plt.imshow(mask_green)
# plt.show()

mask = np.zeros(img.shape, dtype=np.float32)
mask[mask_yellow==255] = [0, 1, 1]
mask[mask_green==255] = [0, 1, 0]
mask[mask_red==255] = [0, 0, 1]
imshow(mask)
imshow(img)
```
