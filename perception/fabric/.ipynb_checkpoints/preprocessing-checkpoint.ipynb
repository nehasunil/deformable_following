```python
import os
import matplotlib.pyplot as plt
from util.segmentation import segmentation

root_dir = "/home/gelsight/Code/Fabric/src/data/imgs/0625_10k"
filename = [f for f in os.listdir(root_dir) if f.startswith("color")]
print(filename)
```

```python
import cv2
import numpy as np
from tqdm import tqdm    
import skimage.measure

W, H = 163, 163
output_dir = "/home/gelsight/Code/Fabric/src/data/imgs/0625_10k_100x100"
os.makedirs(output_dir, exist_ok=True)
for i in tqdm(range(len(filename))):
    fn = filename[i]
    
    img_path = os.path.join(root_dir, fn)
    img = cv2.imread(img_path)
    img_small = cv2.resize(img, (W, H))
    cv2.imwrite(os.path.join(output_dir, fn), img_small)
    
    
    imidx = fn.split("_")[1].replace(".jpg", "")
    
    label_red_path = os.path.join(root_dir, "label_red_"+imidx+".png")
    label_yellow_path = os.path.join(root_dir, "label_yellow_"+imidx+".png")
    label_green_path = os.path.join(root_dir, "label_green_"+imidx+".png")
    
    label_red = cv2.imread(label_red_path)
    label_red_small = cv2.resize(label_red, (W, H), interpolation = cv2.INTER_NEAREST)
    cv2.imwrite(os.path.join(output_dir, "label_red_"+imidx+".png"), label_red_small[:,:,0])
    
    
    label_green = cv2.imread(label_green_path)
    label_green_small = cv2.resize(label_green, (W, H), interpolation = cv2.INTER_NEAREST)
    cv2.imwrite(os.path.join(output_dir, "label_green_"+imidx+".png"), label_green_small[:,:,0])
    
    
    label_yellow = cv2.imread(label_yellow_path)
    label_yellow_small = cv2.resize(label_yellow, (W, H), interpolation = cv2.INTER_NEAREST)
    cv2.imwrite(os.path.join(output_dir, "label_yellow_"+imidx+".png"), label_yellow_small[:,:,0])
    
    depth_path = os.path.join(root_dir, "depth_"+imidx+".npy")
    depth = np.load(depth_path)
    depth_small = skimage.measure.block_reduce(depth, (4,4), np.mean)
    np.save(os.path.join(output_dir, "depth_"+imidx+".npy"), depth_small)
    
    
    depth_img_path = os.path.join(root_dir, "depth_"+imidx+".jpg")
    depth_img = cv2.imread(depth_img_path)
    depth_img_small = cv2.resize(depth_img, (W, H))
    cv2.imwrite(os.path.join(output_dir, "depth_"+imidx+".jpg"), depth_img_small)
```
